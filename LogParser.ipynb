```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
logfile = Path("3237.A05867.19.6.5.HourlyLog.csv")
df = pd.read_csv(logfile, skiprows=2)
df.columns
```

```python
def calc_interval_usage(df, timestamp, total_corrected, total_uncorrected):
    df = df.copy(deep=True)
    df.sort_values(by=[timestamp], ascending=False, inplace=True)
    df['corr interval_usage'] = df[total_corrected].diff(periods=-1)
    df['uncorr_interval_usage'] = df[total_uncorrected].diff(periods=-1)
    df.dropna(inplace=True)
    return df

calc_interval_usage(df, 'time', 'Total Forward Corrected Volume(CF)', 'Total Forward Uncorrected Volume(CF)')
```
